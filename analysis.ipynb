# Analyse von Geschlechterklischees in Berufen

Diese Analyse untersucht die Geschlechterverteilung in den Berufen Ceo und Büroangestellten auf Basis von simulierten Wikidata Daten.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ceo.csv")
```

```python
print(data)

plt.bar(data["gender"], data["count"])
plt.title("Geschlechterverteilung CEO in Deutschland")
plt.xlabel("Geschlecht")
plt.ylabel("Anzahl")

plt.show()
```

## Interpretation

Die Abbildung zeigt eine stark unausgeglichene Geschlechterverteilung im CEO-Beruf in Deutschland. Die Anzahl männlicher CEOs ist deutlich höher als die Anzahl weiblicher CEOs. Frauen sind in dieser Führungsposition stark unterrepräsentiert.

Diese Verteilung spiegelt bekannte gesellschaftliche Strukturen wider, in denen Führungs- und Machtpositionen häufiger von Männern besetzt sind. Der CEO-Beruf gilt als prestigeträchtig und öffentlich sichtbar, wodurch insbesondere männliche Personen stärker in Wikidata erfasst werden.

Zudem ist zu beachten, dass Wikidata keine vollständige Abbildung der gesamten Bevölkerung darstellt, sondern vor allem Personen mit öffentlicher Relevanz enthält. Dadurch kann sich die Ungleichverteilung zusätzlich verstärken, da weniger sichtbarere Gruppen seltener dokumentiert werden.

Insgesamt deutet die Analyse darauf hin, dass Wikidata bestehende Geschlechterungleichheiten nicht nur abbildet, sondern möglicherweise auch verstärkt.

# Analyse von Geschlechterklischees in Berufen

Diese Analyse untersucht die Geschlechterverteilung in den Berufen Arzt und Pflege auf Basis von simulierten Wikidata Daten.

```python
import pandas as pd

data = pd.read_csv("Arzt.csv")
data

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=data, x="occupation", y="count", hue="gender")

plt.title("Geschlechterverteilung nach Beruf (Wikidata)")
plt.xlabel("Beruf")
plt.ylabel("Anzahl")

plt.show()
```

## Interpretation

Die Ergebnisse zeigen eine starke geschlechtsspezifische Segregation in den untersuchten Berufen.

Im Arztberuf sind männliche Personen mit über 90 % deutlich überrepräsentiert. Im Gegensatz dazu zeigt der Pflegeberuf eine starke Dominanz weiblicher Personen mit über 80 %.

Diese Verteilung entspricht klassischen gesellschaftlichen Geschlechterstereotypen, in denen hoch angesehene und akademische Berufe stärker mit Männern assoziiert werden, während fürsorgliche und unterstützende Tätigkeiten häufig Frauen zugeschrieben werden.

Darüber hinaus ist zu beachten, dass es sich um Wikidata-Daten handelt, die nicht die tatsächliche Bevölkerungsstruktur abbilden, sondern die Sichtbarkeit bestimmter Personen widerspiegeln. Dadurch können bestehende Ungleichheiten zusätzlich verstärkt werden.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

arzt = pd.read_csv("arzt.csv")

arzt["count"] = pd.to_numeric(arzt["count"])

arzt["percentage"] = arzt["count"] / arzt["count"].sum() * 100

sns.barplot(data=arzt, x="occupation", y="percentage", hue="gender")

plt.title("Anteile der Berufe an der Gesamtdatenbasis")
plt.xlabel("Beruf")
plt.ylabel("Anteil in %")

plt.show()
```

## Interpretation der prozentualen Darstellung

Die Darstellung zeigt, dass männliche Ärzte den größten Anteil an den untersuchten Wikidata-Einträgen ausmachen. Weibliche Ärzte sind deutlich weniger vertreten. 

Auch im Pflegebereich zeigt sich ein Ungleichgewicht, allerdings in umgekehrter Richtung: Pflegeberufe sind stärker weiblich geprägt.

Diese Verteilung deutet darauf hin, dass sowohl gesellschaftliche Geschlechterrollen als auch die Auswahlkriterien von Wikidata einen Einfluss auf die dargestellten Daten haben.

## Unterschied zwischen absoluter und relativer Darstellung

Während die absoluten Zahlen die tatsächliche Anzahl der Einträge zeigen, ermöglicht die prozentuale Darstellung einen besseren Vergleich zwischen den Berufen.

Da die Gesamtanzahl der Einträge pro Beruf stark variiert, ist die Prozentdarstellung besonders hilfreich, um strukturelle Unterschiede sichtbar zu machen.
