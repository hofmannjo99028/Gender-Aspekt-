```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "occupation": ["CEO", "CEO"],
    "gender": ["male", "female"],
    "count": [102, 9]
})

data["percentage"] = data["count"] / data["count"].sum() * 100

print(data)

plt.bar(data["gender"], data["count"])
plt.title("Geschlechterverteilung CEO in Deutschland")
plt.xlabel("Geschlecht")
plt.ylabel("Anzahl")

plt.show()
```
